Importando a base de dados (dataframe) como "df", e a arazenando-a em uma variável

* read_csv - Lê o dataframe especificado. É necessário especificar o tipo de documento que será lido (csv, json, etc)

* sep = ";" - Em alguns arquivos csv, é necessário especificar o tipo de separador (vírgula ou ponto e vírgula)

* parse_dates=[""] - Converte um tipo de dado qualquer no tipo "Date" 
 

```python
df = pd.read_csv("/content/drive/MyDrive/ocorrencia.csv", sep=";", parse_dates=["ocorrencia_dia", "divulgacao_dia_publicacao"], dayfirst=True)
```

Para substinuir campos com certo valor por um outro valor

* Replace é mais comumente utilizado para substituir dados incoerêntes, classificando-os como vazios.
* pd.NA - Biblioteca do pandas que insere valores nulos no campo

```python
df.replace(["****", "***", "####", "NULL"], pd.NA, inplace=True)
df.head(10)
```
