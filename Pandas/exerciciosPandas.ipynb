```python
import pandas as pd
df = pd.read_csv("/content/drive/MyDrive/ocorrencia.csv", sep=";", parse_dates=["ocorrencia_dia", "divulgacao_dia_publicacao"], dayfirst=True)
```

```python
#Criar um backup para as colunas "UF", "cidade" e "aerodromo"

df["backup_uf"] = df.ocorrencia_uf
df["backup_cidade"] = df.ocorrencia_cidade
df["backup_aerodromo"] = df.ocorrencia_aerodromo
df.head(5)
```

```python
#Normalizar aerodromo. Trocar "***" para MODF
#df.loc[df.nome_coluna == "valor", ["nome_coluna"]] = novo_valor

df.loc[df.ocorrencia_aerodromo == "****", ["ocorrencia_aerodromo"]] = "MODF"
df
```

```python
#Onde existir a cidade "Rio de Janeiro" trocar a classificação de acidente para "tragédia"
#df.loc[df.nome_coluna == "valor", ["nome_coluna"]] = novo_valor

df.loc[df.ocorrencia_cidade == "RIO DE JANEIRO", ["ocorrencia_classificacao"]] = "TRAGÉDIA"
df
```

```python
#Onde existir o aerodromo "sbpk" troque a cidade para "especial"
#df.loc[df.nome_coluna == "valor", ["nome_coluna"]] = novo_valor

df.loc[df.ocorrencia_aerodromo== "SBPK", ["ocorrencia_cidade"]] = "ESPECIAL"
df
```
