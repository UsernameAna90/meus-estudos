```python
#Resolução do "Index out of Range":

listinha = ["Ana","Iris,","Lucas","Mari"]

num = input("Digite um nome para remover da lista: ")

print("Antes da remoção: ",listinha)

for i in range(len(listinha) + 1):
  if num == listinha[i]:
    del listinha[i]
    break #Bastava colocar um break

print("Depois da Remoção: ",listinha)
```

```python
#Os dicionários armazenam diversos elementos organizados na forma de chave-valor: #:.<35

#Criando um dicionário:
#nome_dicionário = {chave:valor1, chave:valor2, chave:valorN}

dicio = {1:"Refrigerante", 2:"Suco Natural", 3:"Energético", 4:"Água Mineral", 5:"Água com Gás", 6:"Água de Coco"}

print(dicio)

dicio[7] = "Cerveja" #Adciona um novo item

print(dicio)

dicio.pop(7) #Excluindo um item

print(dicio)
```

```python
#Criando dicionário com o comando "dict":
#nome_dicionário = dict(valor=chave1, valor=chave2, valor=chaveN)

dicio2 = dict(Refrigerante=1, Suco=2, Energético=3, Água=4) #O valor não precisa estar entre aspas

print(dicio2)
```

```python
#Criando dicionário com o "zip" para concatenar "chave-valor" ao "dict":
#nome_dicionário = dict(zip([valor1, valor2, valorN],[chave1, chave2, chaveN]))

dicio3 = dict(zip(["Refrigerante", "Suco Natural", "Energético", "Água Mineral"],[1, 2, 3, 4]))

print(dicio3)

print(dicio3['Suco Natural']) #Imprimindo um valor específico
```

```python
#Dicionario com dicionarios dentro:

dicionario_dicio = {
    "pessoa1":{"nome":"Ana", "idade":"21", "sexo":"Feminino"}, #Pessoa1 seria a chave, e os itens no colchete seriam o valor
    "pessoa2":{"nome":"Iris", "idade":"X", "sexo":"Feminino"}  #Pessoa2 seria a chave, e os itens no colchete seriam o valor
    }

print(dicionario_dicio) #Printa todos os itens

for i, j in dicionario_dicio.items(): #Printa os itens de cada dicionario, um por um
  for chave, valor in j.items():
    print(chave, valor)
```

```python
#Dicionario com lista de valores:

dicionario_lista = [
    {"Nome":"Ana", "Idade":21},
    {"Nome":"Mari", "Idade":20},
    {"Nome":"Iris", "Idade":"X"}
    ]

print(dicionario_lista)

dicionario_lista[1] = {"nome":"Iris", "Idade":"X"} #Subistituindo um grupo de valores

for i in dicionario_lista: #Printa os itens de cada dicionario, um por um
  for chave, valor in i.items():
    print(chave, valor)
```

```python
#Dicionario simples:

dicionario_simples = {"nome":"Ana", "Idade":21}, {"nome":"Mariana", "Idade":20}

print(*dicionario_simples)

print(dicionario_simples[1]["nome"]) #Printa um valor numa posicao especifica
print(dicionario_simples[0])

for i in dicionario_simples: #Printa os itens de cada dicionario, um por um
  for chave, valor in i.items():
    print(chave, valor)
```
