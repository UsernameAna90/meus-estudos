```python
#Em uma turma há 10 alunos. Cada aluno tem 2 notas. 
#Um professor precisa calcular a média das duas notas de cada aluno. Crie um programa que resolva este problema:

i = 0
n1 = 0
n2 = 0
ME = 0

for i in range(1, 3, 1):
  n1 = float(input(f"Digite a primeira nota do {i}° aluno: "))
  while (n1 < 0) or (n1 > 10):
     n1 = float(input(f"Nota inválida! O valor deve estar entre 0 e 10.\nPor favor, gigite a primeira nota do {i}° aluno: "))
  n2 = float(input(f"Digite a segunda nota do {i}° aluno: "))
  while (n2 < 0) or (n2 > 10):
    n2 = float(input(f"Nota inválida! O valor deve estar entre 0 e 10.\n Por favor, gigite a primeira nota do {i}° aluno: "))

  ME = (n1 + n2) / 2

  print(f"{i}° aluno:\nNota 1 = {n1}\nNota 2 = {n2}\nMedia = {ME:.2f}")
```

```python
#Utilizando a estrutura de repetição for, faça um programa em Python que receba 10 números,
#conte quantos deles estão no intervalo [10,20] e quantos deles estão fora do intervalo, escrevendo estas informações.


i = 0
num = 0
no_in = 0
fora_in = 0

for i in range(1, 11, 1):
  num = int(input(f"Digite o {i}° numero: "))
  if (num >= 10) and (num <= 20):
    no_in += 1
  else:
    fora_in += 1

print(f"{no_in} numeros estavam dentro do intervalo, e {fora_in} numeros estavam fora do intervalo!")
```

```python
#Escreva um programa que leia um valor correspondente ao número de jogadores de um time de vôlei. 
#O programa deverá ler uma altura para cada um dos jogadores e, ao final, informar a altura média do time

i = 0
num = 0
altura = 0
ME = 0
cont = 0

num = int(input("Informe quantos jogadores tem o time de volei: "))
while (num <= 6) or (num >= 12):
  num = int(input("Quantidade invalida! Um time de volei possui seis jogadores em campo e no minimo seis na reserva.\nPor favor, informe quantos jogadores tem o time de volei: "))

for i in range(1, num + 1, 1):
  altura = float(input(f"Informe a altura do {i}° jogador: "))
  ME += altura
  cont += 1

ME = ME / cont 

print(f"A media da altura dos jogadores do time de volei e de {ME:.2f}")
```

```python
#Crie um programa em Python que leia N números (N digitado pelo usuário) 
#e mostre na tela a soma e a média de todos os números lidos.

i = 0
num = 0
num2 = 0
soma = 0
ME = 0

num = int(input("Escolha uma quantidade de numeros para serem lidos: "))

for i in range(1, num + 1, 1):
  num2 = float(input(f"Digito o {i}° numero: "))
  soma += num2
  ME += 1

ME = soma / ME

print(f"A soma dos numeros lidos e igual a {soma:.2f}, e a media e igual a {ME:.2f}")
```

```python
#Faça um programa que receba um número positivo digitado pelo usuário e calcule a soma de todos os números de 1 até ao número digitado. 
#Por exemplo, se o usuário inseriu 4, a saída deve ser 10 (1+2+3+4=10)

i = 0
soma = 0

num = int(input("Digite um numero positivo: "))
while num < 0:
  num = int(input("Numero invalido! O numero deve ser positivo.\nPor favor, digite um numero positivo: "))

for i in range(1, num + 1, 1):
  soma += i

print(f"A soma de todos os numeros de 1 a {num} e igual a {soma}!")
```

```python
#Escreva um programa em Python para encontrar o fatorial de qualquer número.
#FAT 5 = 5 x 4 x 3 x 2 x 1 = 120

i = 0
fat = 1

num = int(input("Digite o numero do qual se deseja encontrar o fatorial: "))
while num < 0:
  num = int(input("Numero invalido! Nao existe fatorial de numero negativo.\nPor favor, digite um numero positivo: "))

for i in range(num, 1, -1):
  fat = fat * i

print(f"O fatorial de {num} e igual a {fat}!")
```

```python
#Escreva um programa em Python que leia um número inteiro positivo e diga se o número informado é primo
 
i = 0
cont = 0
 
num = int(input("Digite um numero inteiro positivo para saber se ele e primo: "))
while num < 0:
  num = int(input("Numero invalido! O numero deve ser positivo.\nPor favor, digite um numero positivo: "))

for i in range(1, num + 1):
  if num % i == 0:
    cont += 1

if cont == 2:
  print(f"O numero {num} e primo!")
else:
  print(f"O numero {num} nao e primo, e possui {cont} divisores!")
```
