```python
num = 50
cont = 0

while num <= 220:
  if num % 2 == 0:
    cont += 1
  num += 1

print(cont)
```

```python
#Faça um algoritmo que leia 10 números entre 100 e 500 e mostre na tela o maior e o menor número lido:

i = 1
num = 0
num_maior = 0
num_menor = 501

while i <= 10:
  num = int(input(f"Escreva {i}° número entre 100 e 500: "))
  while (num < 100) or (num > 500):
    num = int(input(f"O numero deve estar entre 100 e 500!\nPor favor, escreva o {i}° número entre 100 e 500: "))
  if num > num_maior:
    num_maior = num
  if num < num_menor:
    num_menor = num
  i += 1

print(f"O menor numero digitado foi {num_menor} e o maior numero digitado foi {num_maior}.")
```

```python
#Faça um programa que apresente na tela os 10 primeiros números negativos em ordem inversa

i = 10
cont = 0

while i > 0:
  cont = i - (2 * i)
  print(cont)
  i -= 1
```

```python
#Faça um programa que obrigue o usuário à digitar um número negativo 
#Mostre o número negativo na tela 

num = int(input("Digite um número negativo: "))

while num >= 0:
  num = int(input(f"O número {num} é positivo!\nPor favor, digite um número negativo: "))

print(f"O número negativo digitado foi {num}!")
```
