```python
#Faça um algoritmo que imprima na tela se um número lido é par ou ímpar:

num = int(input("Digite um número: "))

if (num % 2) == 0:
  print(f"O número {num} é par!")
else:
  print(f"O número {num} é ímpar!") 
```

```python
#Faça um algoritmo que leia dois valores e imprima qual for maior:

num1 = int(input("Digite um numero: "))
num2 = int(input("Digite outro numero: "))

if num1 == num2:
  print(f"Os números {num1} e {num1} são iguais!")
else:
  if num1 > num2:
    print(f"O numero {num1} e maior que o numero {num2}")
  else:
    print(f"O numero {num2} e maior que o numero {num1}")
```

```python
#Um posto está vendendo combustível com a seguinte tabela de descontos:

litros = float(input("Quantos litros você deseja abastecer?\n"))
tipo = input("Você deseja abastecer com alcool ou gasolina?\nDigite A para alcool, e G para gasolina:")
  
if tipo.upper() == "A":
  if litros <= 20:
    preco = litros * 5.2
    preco = preco - ((preco * 3) / 100) 
    print(f"Você receberá um desconto de 3%! Portanto, o total a ser pago é de R${preco:.2f}")
  else:
    preco = litros * 5.2
    preco = preco - ((preco * 5) / 100)
    print(f"Você receberá um desconto de 5%! Portanto, o total a ser pago é de R${preco:.2f}")
elif tipo.upper() == "G":
  if litros <= 20:
    preco = litros * 6.1
    preco = preco - ((preco * 4) / 100) 
    print(f"Você receberá um desconto de 4%! Portanto, o total a ser pago é de R${preco:.2f}")
  else:
    preco = litros * 6.1
    preco = preco - ((preco * 6) / 100)
    print(f"Você receberá um desconto de 6%! Portanto, o total a ser pago é de R${preco:.2f}")
else:
  print("Tipo inválido!")
```

```python
"""
Faça um algoritmo para ler um número que é um código de usuário. 
Caso este código seja diferente de um código armazenado internamente no algoritmo (igual a 1234) deve ser apresentada a mensagem:‘Usuário inválido!’. 
Caso o Código seja correto, deve ser lido outro valor que é a senha. 
Se esta senha estiver incorreta (a certa é 9999) deve ser mostrada a mensagem ‘senha incorreta’. 
Caso a senha esteja correta, deve ser mostrada a mensagem ‘Acesso permitido’.
"""
cod2 = 1234
sen2 = 9999
cod = int(input("Digite seu codigo de usuario: "))

if cod == cod2:
  sen = int(input("Digite sua senha: "))
  if sen == sen2:
    print("Acesso permitido!")
  else:
    print("Senha incorreta!")
else:
  print("Usuario invalido!")
```

```python
"""
Elabore um algoritmo que calcule o que deve ser pago por um produto, considerando o preço normal de etiqueta e a escolha da condição de pagamento. 
Utilize os códigos da tabela a seguir para ler qual acondição de pagamento escolhida e efetuar o cálculo adequado.
Código Condição de pagamento:
1 À vista em dinheiro ou cheque, recebe 10% de desconto
2 À vista no cartão de crédito, recebe 15% de desconto
3 Em duas vezes, dinheiro, preço normal de etiqueta sem juros
4 Em duas vezes, cartão de crédito, preço normal de etiqueta mais juros de 10%
"""

valor = int(input("Digite o valor da sua compra: "))
print(f"Meus parabéns pela sua compra no valor de R${valor:.2f} reais!\nSelecione a forma de pagamento:")
for_pag = int(input("Para pagar à vista, digite 1. Para pagar em duas parcelas, digite 2.\n"))
din_car = int(input("Para pagar no dinheiro, digite 1. Para pagar no cartao, digite 2.\n"))

if (for_pag == 1) and (din_car == 1):
  valor = valor - ((valor * 10) / 100)
  print(f"Você receberá 10% de desconto! Sendo assim, o valor final a ser pago será de R${valor:.2f} reais!")
elif (for_pag == 1) and (din_car == 2):
  valor = valor - ((valor * 15) / 100)
  print(f"Você receberá 15% de desconto! Sendo assim, o valor final a ser pago será de R${valor:.2f} reais!")
elif (for_pag == 2) and (din_car == 1):
  print(f"Você pagará o preço normal de etiqueta sem juros! Sendo assim, o valor final a ser pago será de R${valor:.2f} reais!")
elif (for_pag == 2) and (din_car == 2):
  valor = valor + ((valor * 10) / 100)
  print(f"Você pagará o preço normal de etiqueta com juros de 10%! Sendo assim, o valor final a ser pago será de R${valor:.2f} reais!")
```

```python
"""
Escreva um algoritmo que leia o número de identificação, 
as 3 notas obtidas por um aluno nas 3 verificações,
a média dos exercícios que fazem parte da avaliação, 
e calcule a média de aproveitamento, usando a fórmula:
MA := (nota1 + nota 2 * 2 + nota 3 * 3 + ME)/7
A atribuição dos conceitos obedece a tabela abaixo. 
O algoritmo deve escrever o número do aluno, suas notas, a média dos exercícios, a média de aproveitamento, o conceito correspondente,
a mensagem 'Aprovado' se o conceito for A, B ou C, e 'Reprovado' se o conceito for D ou E.

Média de aproveitamento Conceito

>= 90 A
>= 75 e < 90 B
>= 60 e < 75 C
>= 40 e < 60 D
< 40 E 
"""
cod = int(input("Digite seu numero de identificacao: "))
n1 = float(input("Digite a nota da primeira avaliação: "))
n2 = float(input("Digite a nota da segunda avaliação: "))
n3 = float(input("Digite a nota da terceira avaliação: "))

ME = (n1 + n2 + n3) / 3
MA = (n1 + n2 * 2 + n3 * 3 + ME) / 7

if MA >= 90:
  con = "A"
elif (MA >= 75) and (MA < 90):
  con = "B"
elif (MA >= 60) and (MA < 75):
  con = "C"
elif (MA >= 40) and (MA< 60):
  con = "D"
elif MA < 40:
  con = "F"

if (con == "A") or (con == "B") or (con == "C"):
  res = "aprovado!"
elif (con == "D") or (con == "F"):
  res = "reprovado..."

print(f"Olá aluno(a) numero {cod}! Suas notas foram:\n1° Avaliacao: {n1}\n2° Avaliacao: {n2}\n3° Avaliacao: {n3}")
print(f"Sua média é igual a {ME:.2f}, sendo que a media de aproveitamento e igual a {MA:.2f}.\nSendo assim, seu conceito é igual a {con}.")
print(f"Resultado final: Voce esta {res}")
```
