```python
def funcao_soma(par1, par2):
  res = par1 + par2
  return res
```

```python
x = funcao_soma(5, 10)

print("Esse é o resultado: ",x)
```

```python
def funcao_print(par1):
  print(par1[::-1])
```

```python
listinha = [1, 2, 3]

funcao_print(listinha)
```

```python
def funcao_print2(par1):
  for i in range(len(par1)):
    print(f"Índice {i}: {par1[i]}")
```

```python
lista_nomes = ["Ana","Clara","Andrade","Paiva"]

funcao_print2(lista_nomes)
```
