```python
#Faça um Programa que leia uma lista de 10 números reais e mostre-os na ordem inversa:

numeros = []
i = 0

for i in range(10):
  numeros.append(float(input(f"Digite o {i + 1}° numero real: ")))
  while (numeros[i] < 0) or (numeros[i] %  1 == 0): #Um numero real deve deixar restos quando dividido por 1
    numeros[i] = float(input("Numero invalido! Por favor, digite um numero real positivo:"))

print(numeros[::-1])
```

```python
#Faça um Programa que leia 4 notas de alunos e armazene-as dentro de uma lista. 
#Em seguida mostre as notas e a média das notas na tela:

notas = []
n = 4
media = 0
soma = 0

for i in range(n):
  notas.append(float(input(f"Digite a {i + 1}ª nota: "))) 
  while (notas[i] < 0) or (notas[i] > 10):
    notas[i] = float(input(f"Nota invalida! A nota deve estar entre 0 a 10.\nPor favor, digite a {i + 1}ª nota novamente: "))
  soma += notas[i]

media = soma / len(notas)
print(f"O aluno obteve as seguintes notas: \n{notas}\nMedia final: {media:.2f} pontos!")
```

```python
#Faça um Programa que leia 20 números inteiros e armazene-os numa lista. 
#Em seguida, crie mais 2 listas, uma para armazenar os números pares e outra para armazenar os números impares. 
#Ao final, Imprima as três listas.

num = []
num_par = []
num_impar = []

for i in range(20):
  num.append(int(input(f"Digite o {i + 1}° numero: ")))
  while (num[i] < 0):
    num[i] = int(input(f"Numero invalido! Por favor, digite um {i + 1}° numero positivo:"))
  if num[i] % 2 == 0:
    num_par.append(num[i])
  else:
    num_impar.append(num[i])

print(f"Lista final {num}\nLista com numeros pares {num_par}\nLista com numeros impares {num_impar}")
```

```python
#Faça um Programa que peça as quatro notas de 10 alunos junto com o seu nome.  
#Calcule e armazene numa lista a média de cada aluno, bem como em outra lista armazene o nome do aluno. 
#Ao final imprima o número de alunos com média maior ou igual a 7.0 junto com o seu nome.

nomes = []
medias = []
nota = 0
soma = 0

for i in range(10):
  nomes.append(input(f"Aluno {i + 1} - Digite seu nome: "))
  for j in range(4):
    nota = float(input(f"Digite a {j + 1}ª nota: "))
    while (nota < 0) or (nota > 10):
      nota = float(input(f"Nota invalida! Por favor, digite a {j + 1}ª nota novamente: "))
    soma += nota
  medias.append(soma/4)
  soma = 0

print(f"Alunos que foram aprovados:")

for i in range(10):
  if medias[i] >= 7:
    print(f"Nome: {nomes[i]}. Media: {medias[i]}")
```

```python
#Faça um programa que leia o nome, a idade e a altura de 10 alunos da SoulCode. 
#Em seguida determine quantos alunos com mais de 13 anos possuem altura inferior à média de altura desses alunos

lista_n = []
lista_i = []
lista_a = []
soma = 0
cont = 0
ME = 0
rg = 10

for i in range(rg):
  print(f"Olá SoulCoder numero {i + 1}!")
  lista_n.append(input("Informe seu nome: "))
  lista_i.append(int(input("Informe sua idade: ")))
  while (lista_i[i] <= 0) or (lista_i[i] >= 100):
    lista_i[i] = int(input(f"Idade invalida! Por favor, informe sua idade novamente: "))
  lista_a.append(float(input("Informe sua altura: ")))
  while (lista_a[i] < 0) or (lista_a[i] > 2.5):
    lista_a[i] = float(input(f"Altura invalida! Por favor, informe sua altura novamente: "))
  soma += lista_a[i]

ME = soma / len(lista_a)
print(f"A altura media dos alunos da SoulCode e de {ME:.2f}. Os seguintes alunos possuem altura abaixo da media:")

for i in range(len(lista_i)):
  if (lista_i[i] > 13) and (lista_a[i] < ME):
    print(f"Aluno(a) {lista_n[i]}, {lista_i[i]} anos de idade, e altura igual a {lista_a[i]}.")
    cont += 1

print(f"Total: {cont} alunos possuem altura abaixo da media!")
```
