```python
#Escreva um programa que leia as medidas dos lados de um triângulo e escreva se ele é Equilátero, Isósceles ou Escaleno:

lado1 = float(input("Digite a primeira medida do lado do triângulo: "))
lado2 = float(input("Digite a segunda medida do lado do triângulo: "))
lado3 = float(input("Digite a terceira medida do lado do triângulo: "))

if (lado1 <= 0) or (lado2 <= 0) or (lado3 <= 0):
  print("Medida digitada invalida!")
elif (lado1 == lado2 != lado3) or (lado2 == lado3 != lado1) or (lado3 == lado1 != lado2):
  print("É um triângulo isóceles!")
elif (lado1 == lado2) and (lado2 == lado3):
  print("É um triângulo equilátero!")
else:
  print("É um triângulo escaleno!")
```

```python
#Desenvolva um algoritmo em Python para determinar o Índice de Massa Corporal (IMC) de uma pessoa:

peso = float(input("Informe seu peso em Kg: "))
h = float(input("informe sua altura em metros: "))

IMC = peso / (h ** 2)

if IMC < 16:
  print(f"Seu IMC e de {IMC:.2f}. Magreza grave!")
elif IMC >= 16 and IMC < 17:
  print(f"Seu IMC e de {IMC:.2f}. Magreza moderada!")
elif IMC >= 17 and IMC < 18.5:
  print(f"Seu IMC e de {IMC:.2f}. Magreza leve!")
elif IMC >= 18.5 and IMC < 25:
  print(f"Seu IMC e de {IMC:.2f}. Saudavel!")
elif IMC >= 25 and IMC < 30:
  print(f"Seu IMC e de {IMC:.2f}. Sobrepeso!")
elif IMC >= 30 and IMC < 35:
  print(f"Seu IMC e de {IMC:.2f}. Obesidade grau I!")
elif IMC >= 35 and IMC < 40:
  print(f"Seu IMC e de {IMC:.2f}. Obesidade grau II (severa)!")
else:
  print(f"Seu IMC e de {IMC:.2f}. Obesidade grau III (morbida)!")
```

```python
#Escreva um programa para ler o ano de nascimento de uma pessoa e escrever uma mensagem que diga se ela poderá ou não votar este ano

ano = int(input("Digite o ano em que você nasceu:"))

ano_atual = 2021
idade = ano_atual - ano

if idade < 16:
  print(f"Você tem {idade} anos, não poderá votar nesse ano!")
elif idade >= 16 and idade < 18:
  print(f"Você tem {idade} anos, não é maior de idade mas já pode votar nesse ano!")
else:
  print(f"Você tem {idade} anos, já é maior de idade e deve votar esse ano!")
```

```python
#Faça um programa que pede duas notas de um aluno. Em seguida ele deve calcular a média do aluno e dar o seguinte resultado:

n1 = float(input("Digite sua primeira nota: "))
n2 = float(input("Digite sua segundaa nota: "))

ME = (n1 + n2) / 2

if ME == 10:
  print(f"Sua media foi de {ME:.2f}. Aprovado com distincao!")
elif (ME < 10) and (ME >= 7):
   print(f"Sua media foi de {ME:.2f}. Aprovado!")
elif (ME < 7) and (ME >= 0):
   print(f"Sua media foi de {ME:.2f}. Reprovado!")
else:
  print("Notas inválidas!")
```

```python
"""
Faça um Programa que pergunte em que turno você estuda. Peça para digitar M-matutino ou V-Vespertino ou N- Noturno. 
Imprima a mensagem "Bom Dia!", "Boa Tarde!" ou "Boa Noite!" ou "Valor Inválido!", conforme o caso.
"""

turno = input("Em que turno voce estuda?\nDigite M para matutino, V para vespertino, e N para noturno: ")

if turno.upper() == "M":
  print("Bom Dia!")
elif turno.upper() == "V":
  print("Boa Tarde!")
elif turno.upper() == "N":
  print("Boa Noite!")
else:
  print("Turno invalido!")
```

```python
"""
As Organizações Tabajara resolveram dar um aumento de salário aos seus colaboradores,
e lhe contrataram para desenvolver o programa que calculará os reajustes.

Faça um programa que recebe o salário de um colaborador e o reajuste segundo o seguinte critério, baseado no salário atual:

    salários até R$ 280,00 (incluindo) : aumento de 20%
    salários entre R$ 280,00 e R$ 700,00 : aumento de 15%
    salários entre R$ 700,00 e R$ 1500,00 : aumento de 10%
    salários de R$ 1500,00 em diante : aumento de 5%.
    
    Após o aumento ser realizado, informe na tela:
    o salário antes do reajuste;
    o percentual de aumento aplicado;
    o valor do aumento;
    o novo salário, após o aumento.
"""

sal = float(input("Informe o valor do seu salario: "))

if sal <= 280:
  per = 20
 
elif (sal > 280) and (sal <= 700):
  per = 15

elif (sal > 700) and (sal <= 1500):
  per = 10

else:
  per = 5

aum = (sal * per) / 100
novo_sal = sal + aum

print(f"Seu salario antes do reajuste era de R${sal:.2f}")
print(f"O percentual de aumento aplicado foi de {per}%, gerando aumento no valor de R${aum:.2f}")
print(f"Seu salario apos o reajuste passa a ser de R${novo_sal:.2f}!")
```
