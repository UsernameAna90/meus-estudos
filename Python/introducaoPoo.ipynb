```python
#Definir a classe
class Pessoa:

  #Definir o metodo da classe
  def __init__(self, Nome, Idade): #Método construtor
    self.Nome = Nome
    self.Idade = Idade

#Instanciar a classe criando um objeto

Pessoa1 = Pessoa("Ana", 21) #Criando uma pessoa com os atributos da classe "Pessoa".

print(f"Olá {Pessoa1.Nome}! Você tem {Pessoa1.Idade} anos de idade!")

Pessoa2 = Pessoa("Sebastian", 23) #Criando outra pessoa com os mesmos atributos, valores diferentes.

print(f"Olá {Pessoa2.Nome}! Você tem {Pessoa2.Idade} anos de idade!")
```

```python
#Definir a classe
class Humano:

  #Definir o metodo da classe
  def Falar(self):
    print("Este humano fala!")

  def Andar(self):
    print("Este humano anda!")

#Instanciar a classe criando um objeto

PessoaUm = Humano() #Este objeto pode usar todos os metodos da classe. Assim, este humano pode andar e falar.
PessoaUm.Falar()
PessoaUm.Andar()
```
