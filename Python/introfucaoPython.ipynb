```python
#Faça um algoritmo que leia dois números e mostre o primeiro elevado ao segundo e o segundo elevado ao primeiro:

num1 = int(input("Digite o primeiro numero:"))
num2 = int(input("Digite o segundo número:"))

print("O primeiro numero elevado ao segundo:", num1 ** num2)
print("O segundo nummero elevado ao primeiro:", num2 ** num1)
```

```python
#Faça um algoritmo que leia um salário e também as despesas de uma pessoa e diga a porcentagem que a despesa representa em relação ao salário:

salario = float(input("Digite o valor do seu salario:"))
despesas = float(input("Digite o valor das suas despesas:"))

resultado = f'{(despesas * 100) / salario}%'

print("A porcentagem de despesas em relacao ao seu salario e igual a:",resultado)
```

```python
#Faça um algoritmo que solicite base e altura de um retângulo e informe sua área e perímetro:

altura = float(input("Digite o valor da altura do retânfulo:"))
base = float(input("Digite o valor da base do retântuglo:"))

area = base * altura
perimetro = 2 * (base + altura)

print("A área do retângulo é igual a:", area,"\nE o perimetro é igual a:",perimetro)
```

```python
#Faça um algoritmo que solicite o nome e preço de um produto, processe um aumento de 65%, e informe o novo preço de venda:

nome = input("Digite o nome de um produto qualquer:")
preco = float(input("Digite o preco desse produto:"))

preco = preco + ((preco * 65) / 100)

print("O preco de venda do(a)",nome,"com um aumento de 65% é igual a:",preco)
```

```python
"""
Faça um algoritmo que pergunte quanto a pessoa quer emprestado, 
em quantos meses ela pretende pagar,
e faça o cálculo do empréstimo a juros compostos.
Fórmula: M = C x (1 + juros)^t
"""

C = float(input("Digite o capital desejado para o empréstimo:"))
t = int(input("Digite número de meses que levará para pagar o empréstimo:"))

t = t / 12
M = C * ((1 + 0.02) ** t)

print(f"Após a aplicação dos juros, o valor do montante final a ser pago é de:{M:.2f}")
```
