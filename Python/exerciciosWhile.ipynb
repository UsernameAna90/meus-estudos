```python
#Fazer um programa para encontrar todos os números pares entre 1 e 100:

num = 1

while num <= 100:
  if num % 2 == 0:
    print(f"O numero {num} e par!")
  num = num + 1
```

```python
#Fazer um programa para encontrar todos os múltiplos de 3 de 1 a 50:
#Para saber se um número1 é multiplo de um número2, 
#é preciso encontrar um número3 inteiro que multicado por numero2 retorne o numero1.

num = 1
multi = 0

while num <= 50:
  multi = 0
  while multi <= 50:
    if num == multi * 3:
      print(f"O numero {num} e multiplo de 3! Seu multiplicador e o numero {multi}!")
    multi += 1
  num += 1
```

```python
#Ler um valor inteiro (aceitar somente valores entre 1 e 10) e escrever a tabuada de 1 a 10 do valor lido

cont = 1
resul = 0
num = int(input("Digite um numero inteiro entre 1 e 10: "))

while (num < 1) or (num > 10):
  num = int(input(f"O numero {num} esta fora do intervalo pedido!\nPor favor, digite um numero inteiro entre 1 e 10: "))

while cont <= 10:
  resul = num * cont
  print(f"{num} x {cont} = {resul}")
  cont += 1
```

```python
#Faça um programa que leia 5 idades e mostre na tela a média das idades lidas:

i = 1
idade = 0
ME = 0

while i <= 5:
  idade = float(input(f"Digite aqui a {i}ª idade: "))
  ME += idade
  i += 1

ME = ME / 5

print(f"A media das idades lidas e igual a {ME}")
```

```python
#Ler 10 números e informar na tela quantos números pares foram digitados e quantos números ímpares foram digitados:

i = 1
num = 0
quant_par = 0
quant_impar = 0

while i <= 10:
  num = int(input(f"Digite o {i}° numero: "))
  if num % 2 == 0:
    quant_par += 1
  else:
    quant_impar += 1
  i += 1

print(f"Foram digitados {quant_par} numeros pares e {quant_impar} numeros impares!")
```

```python
#Faça um programa que leia 3 notas (de 0 a 10) e informe na tela a média das notas lidas:

i = 1
nt = 0
ME = 0

while i <= 3:
  nt = float(input(f"Digite a {i}ª nota: "))
  while (nt <= 0) or (nt > 10):
    nt = float(input(f"Nota invalida! O valor deve estar entre 0 e 10.\nPor favor, digite a {i}ª nota: "))
  ME += nt 
  i += 1

ME = ME / 3

print(f"A media das notas lidas e igual a {ME:.2f}!")
```
