```python
#Crie um programa que compara duas strings e veja se elas são iguais:

str1 = input("Digite uma palavra: ")
str2 = input("Digite uma palavra: ")

print(str1 == str2)
```

```python
#Faça um programa que leia um nome completo e imprima as 4 primeiras letras do nome:

nome = input("Digite seu nome completo: ")

print(nome[0:4])

#Agora verifique a quantidade de caracteres que contém o nome digitado (não pode contar os espaços em branco):

nome.replace(" ", "")

print(len(nome))
```

```python
#Leia uma frase qualquer e verifique se a pessoa digitou a palavra “cachorro” nessa frase (não importa a maneira como a pessoa escreveu):

frase = input("Digite uma frase: ")

print("cachorro" in frase.lower())
```

```python
#Crie um programa que leia uma string e mostre a string de trás pra frente até o 6º caractere:

str1 = input("Digite uma palavra: ") 

print(str1[-1:-7: -1])
```

```python
#Faça um programa que conte o numero de 1’s que aparecem em um string:

str1 = input("Digite uma número: ")

print(str1.count("1"))
```

```python
#Escreva um programa que substitui as ocorrências de um caractere 0 em uma string por outro caractere 1:

str1 = input("Digite um número: ")

print(str1.replace("0","1"))
```

```python
#Leia um nome e verifique se a primeira letra do nome está com letra maiúscula:

str1 = input("Digite o seu nome: ")

PriLet = str1[0:1]

print("A primeira letra do nome está com letra maiúscula:",PriLet.upper() == PriLet)
```

```python
#Faça um programa que troque todas as ocorrências de uma letra L1 pela letra L2 em uma string. 
#A string e as letras L1 e L2 devem ser fornecidas pelo usuario.

str1 = input("Digite uma palavra: ")
L1 = input("Digite uma letra: ")
L2 = input("Digite outra letra: ")

print(str1.replace(L1, L2))
```

```python
#Faça um programa que receba duas frases distintas e imprima de maneira invertida, trocando as letras A por *:

fr1 = input("Digite uma frase qualquer: ")
fr2 = input("Digite outra frase qualquer: ")

fr1 = fr1.replace("a","*")
fr1 = fr1.replace("A","*")
fr2 = fr2.replace("a","*")
fr2 = fr2.replace("A","*")

print(fr1[::-1])
print(fr2[::-1])
```
