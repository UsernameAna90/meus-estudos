```python
from pyspark.sql import SparkSession
import pyspark.sql.functions as F
from pyspark.sql.types import StructType, StructField, StringType, IntegerType
```

```python
spark = (
    SparkSession.builder
      .master('local')
      .appName('dataframe_withcolumn')
      .config('spark.ui.port', '4050')
      .getOrCreate()
)
```

```python
#Informando um novo cabeçalho
df1 = (spark
       .read
       .format("csv")
       .option("header", "false") #Removeu o cabeçalho original
       .option("inferschema", "false")
       .option("delimiter", ";")
       .load('/content/drive/MyDrive/arquivo_geral.csv').toDF("UM", "DOIS", "TRES", "QUATRO", "CINCO", "SEIS", "SETE")
       #toDF = informou um novo cabeçalho
)
```

```python
df1.show()
df1.printSchema()
```

```python
#Segunda forma de informar um novo cabeçalho
lista_esquema = ['UM', 'DOIS', 'TRES', 'QUATRO', 'CINCO', 'SEIS', 'SETE']
df1 = (spark
       .read
       .format("csv")
       .option("header", "false")
       .option("inferschema", "false")
       .option("delimiter", ";")
       .load('/content/drive/MyDrive/arquivo_geral.csv').toDF(*lista_esquema)
)
df1.show()
```

```python
#Terceira maneira - Utilizando o Select com Fcol e Alias
df2 = (spark
       .read
       .format("csv")
       .option("header", "false")
       .option("inferschema", "false")
       .option("delimiter", ";")
       .load('/content/drive/MyDrive/arquivo_geral.csv')
)

df2 = df2.select(F.col("_c0").alias("um"),
                 F.col("_c1").alias("dois"),
                 F.col("_c2").alias("tres"),
                 F.col("_c3").alias("quatro"),
                 F.col("_c4").alias("cinco"),
                 F.col("_c5").alias("seis"),
                 F.col("_c6").alias("sete"),
)

df2.printSchema()
```
