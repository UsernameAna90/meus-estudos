```python
#Codigo para felicitação de aniversário

import time #Importando a biblioteca time

##### Printando uma linha tracejada com o titulo do programa, para fins esteticos #####
print("\n")
time.sleep(3)
print("-" * 49,"Mensagem de Aniversário","-" * 49)
time.sleep(2)
print("\n")

##### Lendo as informacoes do aniversariante a ser homenageado #####
nome = input("Quem está aniversariando?\nR: ")
dataNasc = int(input(f"\nEm que dia a {nome} nasceu?\nR: "))
mesNasc = input("\nEm que mês?\nR: ")
anoNasc = int(input("\nE em que ano?\nR: "))
sign = input(f"\nE o que a {nome} é pra você?\nR: ")

##### Linha para marcar o inicio da mensagem #####
time.sleep(1)
print("\n")
print("-" * 122)
time.sleep(1)

idade = 2021 - anoNasc

##### Imprimindo a mensagem personalizada com as informacoes lidas #####
mensagem = f"""\n
  {nome};

  O dia {dataNasc} de {mesNasc} do ano de {anoNasc}, é pra mim uma benção.
  Pois nesse dia, há {idade} anos atrás, nasceu a minha {sign}!
"""

print(mensagem)
time.sleep(6)
print("  Eu amo você!", end = " ")
time.sleep(2)
print("E algum dia, vou te levar pra viver no Canadá!\n")
time.sleep(2)
print("  Com carinho, Ana.")
time.sleep(2)

##### Linha para marcar o fim da mensagem #####
print("\n")
print("-" * 49,"Chuva de Felicitações!","-" * 49)
time.sleep(2)

nome = nome.upper()
mensagem = f"FELIZ ANIVERSÁRIO {nome}! "

##### Printando "FELIZ ANIVERSARIO (ANIVERSARIANTE)" ate cobrir a tela #####
for i in range(20):
  for j in range(5):
    print(mensagem, end = " ")
    time.sleep(0.1)
  print("\n")
```
